### Exploratory Data Analysis

We will first begin with basic EDA of our Data sets. Here we will load our datasets, inspect the data they contain and generate summary statistics and visualizations to better understand the data we have on our hand.

```python
import pandas as pd
```

```python
# Loading the data
nvidia_df = pd.read_csv('../data/nvidia_stock.csv')
microsoft_df = pd.read_csv('../data/microsoft_stock.csv')
apple_df = pd.read_csv('../data/apple_stock.csv')
```

```python
# Convert Date Column to datetime

nvidia_df['Date'] = pd.to_datetime(nvidia_df['Date'])
microsoft_df['Date'] = pd.to_datetime(microsoft_df['Date'])
apple_df['Date'] = pd.to_datetime(apple_df['Date'])
```

```python
# Check for missing values

missing_values = {
    'Nvidia': nvidia_df.isnull().sum(),
    'Microsoft': microsoft_df.isnull().sum(),
    'Apple': apple_df.isnull().sum()
}
```

```python
# Displaying information about the datasets

print("NVIDIA Dataset Info:")
print(nvidia_df.info())

# Displaying the first few rows of the datasets
print("\nNVIDIA First 5 Rows:")
print(nvidia_df.head())
```

The dataset seems to have been loaded succesfully, we see all the columns in the dataset and also observe that there are 6562 entries in our dataset. There also seems to be no null values and our attempt at converting the `Date` column has been succesfull.

```python
# Displaying information about the datasets
print("\nApple Dataset Info:")
print(apple_df.info())

# Displaying the first few rows of the datasets
print("\nApple First 5 Rows:")
print(apple_df.head())
```

The dataset seems to have been loaded succesfully, we see all the columns in the dataset and also observe that there are 11139 entries in our dataset. There also seems to be no null values and our attempt at converting the `Date` column has been succesfull.

With our basic steps complete let's move on to visualizing the stock price trends of these datasets.

```python
import matplotlib.pyplot as plt
```

```python
#NVIDIA Stock Price Trend
plt.figure(figsize=(12, 6))
plt.plot(nvidia_df['Date'], nvidia_df['Close'], label='NVIDIA', color='blue', alpha=0.8)
plt.title('NVIDIA Stock Price Trend')
plt.xlabel('Year')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates the historical trend of NVIDIA's stock price, showing a relatively flat movement until around 2016, followed by steady growth and a parabolic rise post-2020. The sharp spike in recent years, particularly in 2023-2024, suggests significant investor confidence, possibly driven by advancements in AI, semiconductor demand, and strong financial performance. The volatility in the latest period indicates fluctuations in market sentiment but maintains an upward trajectory overall. This trend aligns with NVIDIA’s dominant position in GPU and AI technology, making it a key player in the tech sector.

```python
# Microsoft Stock Price Trend
plt.figure(figsize=(12, 6))
plt.plot(microsoft_df["Date"], microsoft_df["Close"], label="Microsoft", color='green', alpha=0.8)
plt.title("Microsoft Stock Price Trend")
plt.xlabel("Year")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates Microsoft’s stock price trend, showcasing steady growth in its early years, followed by a sharp rise and subsequent decline around the dot-com bubble in 2000. From 2000 to 2015, the stock remained relatively stable with gradual appreciation. However, post-2015, Microsoft experienced explosive growth, largely driven by its cloud computing (Azure), AI investments, and enterprise software expansion. The sharp upward trajectory from 2020 onwards aligns with the broader tech boom, digital transformation, and AI-driven advancements. Despite recent volatility, the overall trend remains bullish, reinforcing Microsoft’s dominance as a key player in the global tech industry.

```python
# Apple Stock Price Trend
plt.figure(figsize=(12, 6))
plt.plot(apple_df["Date"], apple_df["Close"], label="Apple", color='red', alpha=0.8)
plt.title("Apple Stock Price Trend")
plt.xlabel("Year")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates Apple’s stock price trend, depicting a prolonged period of minimal movement until the early 2000s, followed by a gradual rise and then a sharp exponential increase post-2010. This growth aligns with Apple’s innovations in consumer technology, including the iPhone, Mac, and expanding services ecosystem. The steep acceleration in stock price from 2020 onwards reflects the company’s strong financial performance, increased reliance on digital products, and dominance in the AI and semiconductor markets. While recent fluctuations indicate market corrections and volatility, the overall trend remains strongly bullish, solidifying Apple’s position as one of the most valuable and influential tech giants.

Let's also visualize trading volume trends of our datasets.

```python
# NVIDIA Trading Volume Trend
plt.figure(figsize=(12, 6))
plt.plot(nvidia_df["Date"], nvidia_df["Volume"], label="NVIDIA", color='blue', alpha=0.8)
plt.title("NVIDIA Trading Volume Trend")
plt.xlabel("Year")
plt.ylabel("Trading Volume")
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates NVIDIA’s trading volume trend over time, showing notable fluctuations in investor activity. The early 2000s exhibit high trading volumes, likely driven by the tech boom and subsequent corrections. A peak around 2003-2004 suggests a period of heightened market interest, followed by a decline in volume. From 2008 to 2016, trading activity remained relatively stable with occasional spikes, possibly due to earnings reports or market events. Post-2016, as NVIDIA’s stock price surged, trading volume saw periodic increases, though not as extreme as the early 2000s. Despite the recent stock price boom, trading volume appears comparatively lower, suggesting a shift toward long-term institutional investment rather than speculative trading.

```python
# Microsoft Trading Volume Trend
plt.figure(figsize=(12, 6))
plt.plot(microsoft_df["Date"], microsoft_df["Volume"], label="Microsoft", color='green', alpha=0.8)
plt.title("Microsoft Trading Volume Trend")
plt.xlabel("Year")
plt.ylabel("Trading Volume")
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates Microsoft’s trading volume trend over time, showing high volatility in the early years, particularly around the 1980s and 1990s, where several spikes suggest periods of intense trading activity, likely due to market excitement and key company milestones. Trading volume remained relatively high until the early 2000s, aligning with the dot-com bubble and its aftermath. However, post-2010, as Microsoft’s stock price steadily increased, trading volume exhibited a gradual decline, indicating a shift toward institutional investment and long-term holdings rather than speculative trading. The lower trading volume in recent years, despite significant stock price growth, suggests that fewer shares are changing hands frequently, reinforcing Microsoft’s stability as a mature and dominant tech leader.

```python
# Apple Trading Volume Trend
plt.figure(figsize=(12, 6))
plt.plot(apple_df["Date"], apple_df["Volume"], label="Apple", color='red', alpha=0.8)
plt.title("Apple Trading Volume Trend")
plt.xlabel("Year")
plt.ylabel("Trading Volume")
plt.legend()
plt.grid(True)
plt.show()
```

The graph illustrates Apple’s trading volume trend, showing significant fluctuations over time. The early years (1980s-1990s) exhibit moderate trading activity, which then spikes dramatically around 2000, likely influenced by the dot-com bubble. A major peak occurs in the early 2000s, with a sharp increase in trading volume, possibly due to pivotal product launches, investor speculation, or market corrections. After 2005, trading volume gradually declines, suggesting a transition from speculative trading to more stable, long-term investment strategies. Despite Apple's exponential stock price growth post-2010, the trading volume has continued to decrease, indicating institutional investors holding onto shares rather than frequent trading. This trend underscores Apple’s maturity as a dominant tech stock, where fewer shares are exchanged despite increasing market valuation.

Next let's generate some heatmaps to see if there exists any correlation between the stock attributes

```python
import seaborn as sns
```

```python
# NVIDIA Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(nvidia_df.drop(columns=['Date']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("NVIDIA Stock Correlation Heatmap")
plt.show()
```

The NVIDIA stock correlation heatmap shows the relationship between different stock metrics. The Open, High, Low, and Close prices all have a perfect positive correlation (1.00) with each other, indicating that these price metrics move in complete synchronization. This is expected since they are closely related and track the stock's daily price fluctuations. However, trading volume has a weak negative correlation (-0.17) with price metrics, suggesting that when stock prices increase, trading volume slightly decreases, and vice versa. This could indicate that higher stock prices attract long-term holders, reducing frequent trading activity. The weak correlation also implies that volume alone may not be a strong predictor of NVIDIA’s stock price movements.

```python
# Microsoft Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(microsoft_df.drop(columns=['Date']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Microsoft Stock Correlation Heatmap")
plt.show()
```

The Microsoft stock correlation heatmap reveals a perfect positive correlation (1.00) between Open, High, Low, and Close prices, indicating that these price metrics move in perfect synchronization. This is typical for stock market data, as these values are inherently linked. However, trading volume shows a moderate negative correlation (-0.36) with price metrics, suggesting that as Microsoft’s stock price rises, trading volume tends to decrease. This could indicate that long-term investors hold onto shares at higher prices, leading to lower trading activity. The negative correlation between volume and price is more pronounced than in NVIDIA’s heatmap, reinforcing the idea that Microsoft’s stock is becoming more institutionally dominated, with reduced short-term trading fluctuations.

```python
# Apple Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(apple_df.drop(columns=['Date']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Apple Stock Correlation Heatmap")
plt.show()
```

The Apple stock correlation heatmap shows a perfect positive correlation (1.00) among the Open, High, Low, and Close prices, indicating that these price variables move in tandem. This is typical for stock data, where daily price fluctuations are strongly interconnected. However, trading volume has a weak negative correlation (-0.27) with stock prices, suggesting that as Apple’s stock price rises, trading volume tends to decrease. This pattern implies that Apple is increasingly held by long-term institutional investors, leading to fewer shares being actively traded. The negative correlation is weaker than in Microsoft’s case but still suggests that higher prices lead to lower trading activity, reinforcing Apple's market stability and investor confidence in holding shares over time.

Let's now detect ouliers in our Dataset by making some boxplots.

```python
# NVIDIA Box Plot
plt.figure(figsize=(8,6))
sns.boxplot(data=nvidia_df[['Open','High','Low','Close']])
plt.title("NVIDIA Stock Price Distribution")
plt.show()
```

The NVIDIA stock price distribution boxplot reveals key insights into price variability and outliers. The Open, High, Low, and Close prices show similar distributions, as expected, given their strong correlation. The median prices are relatively low compared to the upper values, reflecting NVIDIA’s historical price movements before its exponential rise. The plot also highlights numerous outliers, particularly in the upper range, indicating that NVIDIA’s stock experienced significant price spikes over time. The long upper whiskers suggest that recent stock surges have driven prices far beyond historical norms. This distribution aligns with the observed rapid growth in NVIDIA’s stock price post-2020, confirming its increasing market valuation and volatility.

```python
# Microsoft Box Plot
plt.figure(figsize=(8,6))
sns.boxplot(data=microsoft_df[['Open','High','Low','Close']])
plt.title("Microsoft Stock Price Distribution")
plt.show()
```

The presence of numerous outliers, particularly on the upper end, suggests that Microsoft’s stock has experienced sharp price increases, particularly in recent years. This pattern aligns with its steady growth post-2015 and exponential rise in the 2020s, driven by its dominance in cloud computing, AI, and enterprise solutions. The spread in prices highlights increasing volatility over time, but the consistent box sizes suggest relatively stable growth with periodic spikes in valuation.
