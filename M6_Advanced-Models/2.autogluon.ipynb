```python
import pandas as pd
from autogluon.tabular import TabularPredictor
```

```python
# Load preprocessed stock datasets
nvidia_df = pd.read_csv("../data/nvidia_stock_features_selected.csv", index_col=0, parse_dates=True)
microsoft_df = pd.read_csv("../data/microsoft_stock_features_selected.csv", index_col=0, parse_dates=True)
apple_df = pd.read_csv("../data/apple_stock_features_selected.csv", index_col=0, parse_dates=True)

# Set target variable (Close Price)
target = "Close"
```

```python
def train_autogluon(stock_name, df, preset):
    print(f"\nTraining AutoGluon on {stock_name}.")

    train_size = int(0.8*len(df))
    train_data = df[:train_size]
    test_data = df[train_size:]

    predictor = TabularPredictor(label=target, eval_metric='mean_squared_error').fit(train_data, presets = preset)

    predictions = predictor.predict(test_data)

    from sklearn.metrics import mean_squared_error, r2_score
    mse = mean_squared_error(test_data[target], predictions)
    r2 = r2_score(test_data[target], predictions)

    print(f"Mean Squared Error: {mse}")
    print(f"R2 Score: {r2}")

    return predictor, mse, r2
```

```python
nvidia_predictor, nvidia_mse, nvidia_r2 = train_autogluon("NVIDIA",nvidia_df, preset="medium_quality")
```

```python
microsoft_predictor, microsoft_mse, microsoft_r2 = train_autogluon("Microsoft",microsoft_df, preset="medium_quality")
```

```python
apple_predictor, apple_mse, apple_r2 = train_autogluon("Apple",apple_df, preset="medium_quality")
```
